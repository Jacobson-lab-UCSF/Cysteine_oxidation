```python
### comment on this https://pubchem.ncbi.nlm.nih.gov/compound/Cysteine#section=Computed-Properties, polar Accessible area of CYS 64,3A

import numpy as np
import pymol as cmd
from pymol import cmd,stored
import tensorflow as tf

def Cys_environment(cysname, protname):#, dist_cutoff):
    tf.keras.backend.clear_session()
    
    cmd.fetch(protname, "Prot", type="pdb1")
    cmd.split_states("Prot")
    cmd.remove("Prot")
    cmd.delete("Prot")
    Obj=cmd.get_object_list()
    
    list_results=[]

    y=Obj[0]
    
    cmd.split_chains(f"{y}")
    Obj1=cmd.get_object_list(f"{y}_*")
    
    cmd.set('dot_solvent',1)
    cmd.set('dot_density',3)
    
    
    d=0
    #print(len(Obj1))
    while d<len(Obj1):
        cs=Obj1[d]
        CysSelect=(f"{cs} and resname {cysname}") #### Selection for residue
        cmd.select('Cys', CysSelect)
        stored.list=[]
        cmd.iterate("(Cys & name ca)", "stored.list.append((resi,resn))");
#### Cysteines id
        if len(stored.list)==0:
            h=[]
            list_results=[]
            x=[]
            i=[]
        else:
            i=0
            while i<len(stored.list):
                x=stored.list[i][0]
#                h=cmd.get_area(f"{cs} and resi {x} and resname {cysname}")
                RS=cmd.get_sasa_relative(f"{cs} and resi {x} and resname {cysname}")
                for name, h in RS.items():
                    print(x, h)
                    list_results.append(h)
                i=i+1
        list_results2=[*list_results]
        #print(d)
        d=d+1
    cmd.remove('all')
    del d, h, x, i, list_results, CysSelect, Obj
    return list_results2

def extract_env (AAList, Cys_name):
    L=[]
    Lcount=0
    for i in AAList:
        print(i)
        L1=Cys_environment(Cys_name, i)
        Lcount=Lcount + len(L1)
        #flat_list = [item for sublist in L1 for item in sublist]
        L.append(L1)
    LOut=[item for sublist in L for item in sublist]
    tf.keras.backend.clear_session()
    return Lcount, LOut
```
